#### Preprocessing

```python
df1 = df.copy()
```

```python
#drop kolom2 yang tidak relevan dan juga yang memiliki missing value > 80%
df1 =df1.drop(['Username','EnterpriseID','NomorHP','Email','IkutProgramLOP'], axis=1)
df1.info()
```

```python
#mengatasi missing value pada kolom SkorKepuasanPegawai, JumlahKeikutsertaanProjek, JumlahKetidakhadiran, JumlahKeterlambatanSebulanTerakhir
missing = ['SkorKepuasanPegawai', 'JumlahKeikutsertaanProjek', 'JumlahKetidakhadiran', 'JumlahKeterlambatanSebulanTerakhir']
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy= 'median')      
imputer.fit(df1[missing])
df1[missing] = imputer.transform(df1[missing])
df1.info()
```

```python
#mengatasi missing value pada kolom AlasanResign
df1['AlasanResign'].value_counts()
```

```python
#value ui&ux tidak relevan dengan kolom alasan resign, harus diganti dengan yang lebih relevan
df1['AlasanResign'] = df1['AlasanResign'].replace({'Product Design (UI & UX)':'Lainnya'})
```

```python
df1['AlasanResign'].isna().sum()
```

```python
df1[(df1['AlasanResign'].isnull())].head()
```

```python
not_resign = df1[(df1['AlasanResign'].isnull()) & (df1['TanggalResign']=='-')]
not_resign.shape
```

```python
#imputasi missing value kolom AlasanResign dengan masih_bekerja
df1['AlasanResign'].fillna('masih_bekerja', inplace=True)
df1.info()
```
